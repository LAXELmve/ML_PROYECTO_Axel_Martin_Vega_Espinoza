### - De las columnas numéricas, muestre la información estadística relevante (promedio, cuartíles, desviación estándar y coeficiente de variación).

# PARTE II: MANIPULACIÓN Y TRATAMIENTO.

## Paso 1. Tratamiento de datos nulos.

### - Trate los datos nulos acorde a lo que se ha visto previamente en el curso.

```python
df_old = df.copy()
```

<p style="color:blue;">
Dentro del dataset las columnas con mayor cantidad de nulos coinciden en los mismos registros. Considero que tratar de imputar los datos de esos registros no es viable, lo mejor sería eliminar esos registros, ya que tienen gran cantidad de espacios vacíos.
</p>

```python
df = df[df['Drug'].notna()]
```

```python
total_nulos = df.isnull().sum()
plt.figure(figsize=(10, 6))
plt.bar(total_nulos.index, total_nulos.values)
plt.xlabel('Columnas')
plt.xticks(rotation=90)
plt.ylabel('Cantidad de datos nulos')
plt.title('Cantidad de datos nulos paso intermedio')
plt.show()
```

<p style="color:blue;">
Para las demás columnas se realizará imputación con el fin de no perder más información
</p>

```python
for col in df.columns:
        # Si la columna es de tipo objeto, imputar con la moda
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        # Si la columna es numérica, imputar con la media
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

### - Muestre la cantidad de datos nulos antes y después del tratamiento.

```python
total_nulos = df_old.isnull().sum()
plt.figure(figsize=(10, 6))
plt.bar(total_nulos.index, total_nulos.values)
plt.xlabel('Columnas')
plt.xticks(rotation=90)
plt.ylabel('Cantidad de datos nulos')
plt.title('Cantidad de datos nulos antes de tratamiento')
plt.show()
```

```python
total_nulos = df.isnull().sum()
plt.figure(figsize=(10, 6))
plt.bar(total_nulos.index, total_nulos.values)
plt.xlabel('Columnas')
plt.xticks(rotation=90)
plt.ylabel('Cantidad de datos nulos')
plt.title('Cantidad de datos nulos después de tratamiento')
plt.show()
```

## Paso 2. Tratamiento de anomalías.

### - Enfrente los datos extremos y las anomalías.

```python
print(df.shape)
```

```python
df_old2 = df.copy()
```

### - Es libre de utilizar los métodos que prefiera, trate de no perder muchos datos (se tiene un dataset pequeño).

<p style="color:blue;">
Trataré los datos extremos por medio del rango inter cuartil, pero solo para las columnas con mayor correlación.
</p>

```python
columnas = ['Bilirubin']

for columna in columnas:
    q3 = df[columna].quantile(0.75)
    q1 = df[columna].quantile(0.25)
    iqr = q3 - q1
    lsuperior = q3 + (1.5 * iqr)
    linferior = q1 - (1.5 * iqr)
    df = df[(df[columna] > linferior) & (df[columna] < lsuperior)]
```

### - Muestre las distribuciones tratadas antes y después del tratamiento.

```python
for columna in columnas:
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    sns.boxplot(x=df_old2[columna], ax=axs[0])
    axs[0].set_title('Antes de tratar las anomalías')
    
    sns.boxplot(x=df[columna], ax=axs[1])
    axs[1].set_title('Después de tratar las anomalías')
    
    plt.show()
```

# PARTE III: MACHINE LEARNING.

## Paso 1. División de los datos.

### - Divide las columnas en la variable "X" y la variable "y".

```python
X = df.drop(['ID','Status'], axis=1)
y = df['Status']
```

### - Muestree los datos en dos: entrenamiento y pruebas. La proporción de cada muestra queda a decisión suya.

```python
from sklearn.model_selection import train_test_split
```

### - Utilice una semilla para que los resultados puedan ser reproducibles.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=100)
```

## Paso 2. Abordaje mediante modelo simple.

### - Importe un modelo simple de ML, puede ser KNN, Regresión Logística o un árbol de decisión.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=100, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
```

### - Entrene el modelo con los datos y realice las predicciones con la muestra de pruebas.

```python
dtc.fit(x_train, y_train)
```

```python
y_pred = dtc.predict(x_test)
```

### - Muestre los resultados con f1_score, accuracy_score y classification_report.

```python
from sklearn.metrics import f1_score, accuracy_score, classification_report
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("F1 Score:", f1)
print("Accuracy Score:", accuracy)
print("Classification Report:\n", report)
```

### - Analizando el reporte de clasificación, ¿qué tal se desempeñó su modelo?

<p style="color:blue;">
El reporte de clasificación dividió la muestra de pruebas en tres partes de distintas proporciones. En la de mayo proporción se obtuvieron, para las tres métricas, valores por encima del 0.78, siendo el máximo 0.83. Para los otros dos grupos los resultados fueron menores a 0.76, siendo estos últimos los grupos con menores muestras. 
</p>
<p style="color:blue;">
De acuerdo a las puntuaciones generales, el modelo tiene capacidades de predicción aceptables.
</p>

## Paso 4. Ensambles.

### - Utilice el ensamble de VotingClassifier para mejorar el rendimiento.

```python
from sklearn.ensemble import VotingClassifier
```

### - Seleccione al menos 4 modelos simples diferentes y úselos dentro del ensamble (Stacking).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
logistic_reg = LogisticRegression(random_state=100)
tree_clf = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=100, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
rf_clf = RandomForestClassifier(n_estimators=50, random_state=100)
knn_clf = KNeighborsClassifier()
```

```python
eclf = VotingClassifier(estimators=[('logistic', logistic_reg), ('tree', tree_clf), ('rf', rf_clf), ('knn', knn_clf)],voting='soft', n_jobs=2)
```

### - Entrene el meta-modelo y valide su rendimiento con f1_score, accuracy_score y classification_report.

```python
eclf.fit(x_train, y_train)
```

```python
y_pred = eclf.predict(x_test)
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("F1 Score:", f1)
print("Accuracy Score:", accuracy)
print("Classification Report:\n", report)
```
